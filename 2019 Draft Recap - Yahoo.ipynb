```python
import pandas as pd
import numpy as np
import warnings

pd.options.mode.chained_assignment = None
```

```python
#df = pd.read_csv(r'C:\Users\stefa\OneDrive\Documents\Analysis Projects\Fantasy Modeling\Football\2019 Data\Draft Results 2019.csv')
df = pd.read_csv(r'C:\Users\stefa\OneDrive\Documents\Analysis Projects\Fantasy Modeling\Football\2019 Data\Draft Results 2019 - DBRS.csv')
```

```python

pos_list = ['QB','RB','WR','TE','K','D/ST','FLEX']
op_list = ['QB','RB','WR','TE']
flex_list = ['RB','WR','TE']


for pos in pos_list:
    if pos == 'FLEX':
        df[pos+'_pt'] = np.where(df['Pos'].isin(flex_list),df['Points'],0)
    elif pos == 'OP':
        df[pos+'_pt'] = np.where(df['Pos'].isin(op_list),df['Points'],0)
    else:
        df[pos+'_pt'] = np.where(df['Pos']==pos,df['Points'],0)
    df[pos+'_rank'] = df[pos+'_pt'].rank(method='max', ascending=False)
    
df
```

```python
df_teams = pd.DataFrame()
teams = list(pd.unique(df['Owner']))
for team in teams:
    df_temp = df[df['Owner']==team]
    df_temp['taken'] = 0
    for pos in pos_list:
        df_temp[pos+'_rank'] = df_temp[pos+'_pt'].rank(method='max', ascending=False)

        
    st_qb_pts = df_temp[df_temp['QB_rank']==1]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['QB_rank']==1,1,0))
                                
    st_te_pts = df_temp[df_temp['TE_rank']==1]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['TE_rank']==1,1,0))
    
    st_k_pts = df_temp[df_temp['K_rank']==1]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['K_rank']==1,1,0))                              
                                
    st_d_pts = df_temp[df_temp['D/ST_rank']==1]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['D/ST_rank']==1,1,0))
                                
    st_rb_pts = df_temp[df_temp['RB_rank']==1]['Points'].values[0]+df_temp[df_temp['RB_rank']==2]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['RB_rank']==1,1,0))
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['RB_rank']==2,1,0))
                                
                                
    st_wr_pts = df_temp[df_temp['WR_rank']==1]['Points'].values[0]+df_temp[df_temp['WR_rank']==2]['Points'].values[0]
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['WR_rank']==1,1,0))
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['WR_rank']==2,1,0))

    st_flex_pts = df_temp[(df_temp['taken']==0)&(df_temp['Pos'].isin(flex_list))]['Points'].max()
    df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['Points']==st_flex_pts,1,0))
                                
#     st_op_pts = df_temp[(df_temp['taken']==0)&(df_temp['Pos'].isin(op_list))]['Points'].max()
#     df_temp['taken'] = np.where(df_temp['taken']==1,1,np.where(df_temp['Points']==st_op_pts,1,0))

#     try: 
#         qb3_pts = df_temp[df_temp['QB_rank']==3]['Points'].values[0]
#     except: 
#         qb3_pts = 0
    
    bench_pts = df_temp[(df_temp['taken']==1)&(df_temp['Pos'].isin(flex_list))]['Points'].sum()
    
    temp_team = pd.DataFrame(np.column_stack([team,
                                             int(st_qb_pts),
                                             int(st_rb_pts),
                                             int(st_wr_pts),
                                             int(st_te_pts),
                                             int(st_flex_pts),
                                             #int(st_op_pts),
                                             int(st_k_pts),
                                             int(st_d_pts),
                                             #int(qb3_pts),
                                             int(bench_pts)]),columns=['Team','QB','RB','WR','TE','FLEX','K','D','Bench'])
    df_teams = df_teams.append(temp_team)
df_teams
                                
```

```python
df_teams['start_pt'] = df_teams['QB']+df_teams['RB']+df_teams['WR']+df_teams['TE']+df_teams['FLEX']+df_teams['D']+df_teams['K']
df_teams['starter_rank'] = df_teams['start_pt'].rank(method='max', ascending=False)
df_teams['bench_rank'] = df_teams['Bench'].rank(method='max', ascending=False)
df_teams
```
